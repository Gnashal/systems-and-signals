```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import dlti, dimpulse

# Moving average coefficients
b = [1/3, 1/3, 1/3]  # average over 3 samples
a = [1.0, 0.0, 0.0]           # no feedback

# Create discrete-time LTI system
system = dlti(b, a)

# Get impulse response for 10 samples
t, h = dimpulse(system, n=10)
t = np.squeeze(t)
h = np.squeeze(h)

# Plot impulse response
plt.stem(t, h)
plt.xlabel('n (samples)')
plt.ylabel('h[n] (Impulse response)')
plt.title('Impulse Response of a Moving Average System (Order 3)')
plt.grid(True)
plt.show()
```
